```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fetch data from API
url = 'http://api.open-notify.org/astros.json'
response = requests.get(url)

# Check if fetch data successed
if response.status_code == 200:

    # Convert to JSON
    data = response.json()

    # Create DataFrame
    df = pd.DataFrame(data)
    print(df.head(3))
    
else:
    print(f'Error: {response.status_code}')
```

```python
# Extract people data from nested JSON
df = pd.DataFrame(data['people'])
print(df)
```

```python
# Count astronauts by craft
df['craft'].value_counts()

sns.countplot(data=df, x='craft')
plt.title('ISS has 3x more astronauts than Tiangong')
plt.show()
```
