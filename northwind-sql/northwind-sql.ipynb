```python
import sqlite3
import pandas as pd

# Connect to database
conn = sqlite3.connect('database/northwind.db')
cursor = conn.cursor()

# Check all tables - by cursor 
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
print(cursor.fetchall())

# Check all table - by pandas
query = "SELECT * FROM Customers"
df = pd.read_sql_query(query, conn)
df.head()
```

```python
# SELECT specific column
query = '''
SELECT CustomerID, CompanyName, Country
FROM Customers
'''
df = pd.read_sql_query(query, conn)
df
```

```python
# WHERE to filter value
query = '''
SELECT CustomerID, CompanyName, Country
FROM Customers
WHERE Country = 'Germany'
'''
df = pd.read_sql_query(query, conn)
df
```

```python
# ORDER BY to sort value
query = '''
SELECT CustomerID, CompanyName, Country
FROM Customers
WHERE Country = 'Germany'
ORDER BY CompanyName ASC
'''
df = pd.read_sql_query(query, conn)
df
```

```python
# LIMIT to limit rows
query = '''
SELECT CustomerID, CompanyName, Country
FROM Customers
WHERE Country = 'Germany'
ORDER BY CompanyName ASC
LIMIT 3
'''
df = pd.read_sql_query(query, conn)
df
```

```python
# GROUP BY to grouped and conclude value
query = '''
SELECT Country, 
       COUNT(*) as TotalCustomers
FROM Customers
GROUP BY Country
ORDER BY TotalCustomers DESC
'''
df = pd.read_sql_query(query, conn)
df
```

```python
# หาว่าแต่ละ ShipCountry มี
# กี่ออเดอร์ -> COUNT(*)
# ค่าขนส่งรวม -> SUM(Freight)
# ค่าขนส่งเฉลี่ย -> AVG(Freight)
# เรียงจากค่าขนส่งรวมมากไปน้อย

query = '''
SELECT ShipCountry,
       COUNT(*) as TotalOrders,
       SUM(Freight) as TotalFreight,
       AVG(Freight) as AvgFreight
FROM Orders
GROUP BY ShipCountry
ORDER BY TotalFreight DESC
'''

df = pd.read_sql_query(query, conn)
df
```

```python
# JOIN to combine table
query = '''
SELECT Customers.CompanyName, Orders.OrderID, Orders.Freight
FROM Customers
JOIN Orders ON Customers.CustomerID = Orders.CustomerID
'''
df = pd.read_sql_query(query, conn)
df
```

```python
# หาว่าแต่ละ CompanyName 
# มียอด Freight รวมเท่าไหร่ 
# เรียงจากมากไปน้อย 
# แสดงแค่ Top5
query = '''
SELECT CompanyName,
       SUM(Freight) as TotalFreight
FROM Customers
JOIN Orders
ON Customers.CustomerID = Orders.CustomerID
GROUP BY CompanyName
ORDER BY TotalFreight DESC
LIMIT 5
'''
df = pd.read_sql_query(query, conn)
df
```
