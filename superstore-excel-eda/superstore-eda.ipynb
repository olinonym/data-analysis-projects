```python
# Find the best-selling category
top_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

sns.barplot(x=top_category.index, y=top_category.values)
plt.title('Technology has the highest number of orders')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# Find the most profit region
top_region = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)

sns.barplot(x=top_region.index, y=top_region.values)
plt.title('West region has the most profit')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.show()
```

```python
# The most use ship mode
filter_ship_mode = df['Ship Mode'].value_counts()

sns.barplot(x=filter_ship_mode.values, y=filter_ship_mode.index)
plt.title('Standard class is the most use ship mode')
plt.xlabel('Total use of ship mode ')
plt.ylabel('Ship Mode')
plt.show()
```

```python
# Top 10 best-selling products by total sales
top10 = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top10.values, y=top10.index)
plt.title('Top 10 best seller')
plt.xlabel('Count')
plt.show()
```

```python
# Relationship between discount and profit
sns.scatterplot(data=df, x='Discount', y='Profit')
plt.title('Higher discount leads to lower profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```
